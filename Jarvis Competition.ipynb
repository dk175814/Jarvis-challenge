```python
import pandas as pd
import numpy as np
```

### Dataset 

```python
df = pd.read_csv('train.csv')

#df['description'] = df['description'].replace(np.NaN,0)
#df.loc[df['description'] != 0,'description'] = 1

df = df.dropna()

unique1 = set(df['l1'])
print(len(unique1))
print(unique1)
j=1
for i in unique1:
    df.loc[df['l1'] == i,'l1'] = j
    j = j+1
    
unique2 = set(df['l2'])
print(len(unique2))
print(unique2)
j=1
for i in unique2:
    df.loc[df['l2'] == i,'l2'] = j
    j = j+1
    
df1 = df[['region_id','category_id','district_id','listing_num_images','listing_price','l1','l2','demand']]
a = df1[['region_id','category_id','district_id','listing_num_images','listing_price','l1','l2']]

#normalization
#train_stats = a.describe()
#train_stats = train_stats.transpose()
#train_stats
#def norm(z):
#    return (z - train_stats['mean']) / train_stats['std']
#norm_df1 = norm(a)

x=a
#x = norm_df1[['region_id','category_id','listing_num_images','listing_price','l1','l2']]
y = df1['demand']

print(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
print(x_train)
```

### Training

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=75, random_state=0)
model.fit(x_train, y_train)
train_score = model.score(x_train, y_train)
predicted = model.predict(x_train)
mse = np.sum(((predicted - y_train)**2))/347015

print(train_score)
print(mse)
```
